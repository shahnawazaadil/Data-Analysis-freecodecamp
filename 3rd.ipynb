```python
data[~(data['ap_lo']<=data['ap_hi'])]
```

```python
heat_map=data[(data['ap_lo']<=data['ap_hi'])&
     (data['height'] >= data['height'].quantile(0.025))&
    (data['height'] <= data['height'].quantile(0.975))&
    (data['weight'] >= data['weight'].quantile(0.025))&
    (data['weight'] <= data['weight'].quantile(0.975))]
```

```python
corr=heat_map.corr()
```

```python
mask=np.triu(corr)
```

```python
sns.heatmap(corr,mask=mask,annot=True,fmt='0.1f',square=True)
```
